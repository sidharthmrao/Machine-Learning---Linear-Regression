```
import random
import matplotlib.pyplot as plt
from google.colab import files

#Upload data.train file to Colab
files.upload()

#Create Data Sets X and Y

dataX = [] #List of Data Points for X Values
dataY = [] #List of Data Points for Y Values


#Import Values for dataX and dataY from data.train file

for line in open("data.train", "r"):
    values = line.split(",")
    dataX.append(float(values[0]))
    dataY.append(float(values[1]))



#Loss Function: Calculates the Mean Squared Error of the Line Created vs the Points Given

def loss(a, b, x, y):
  sum = 0
  for i in range(len(x)):
    sum+= (a*x[i] + b - y[i])**2

  avg = (1/len(x)) * sum
  return avg

#aPrime Function: Gives the Derivative of the Loss with respect to 'a'; used to change 'a'.

def aPrime(a, b, x, y):

  sum = 0
  for i in range(len(x)):
    sum+= x[i]*(y[i] - a*x[i] - b)

  aPrime = (2/len(x)) * sum
  return aPrime

#bPrime Function: Gives the Derivative of the Loss with respect to 'b'; used to change 'b'.

def bPrime(a, b, x, y):

  sum = 0
  for i in range(len(x)):
    sum+= (y[i] - a*x[i] - b)
    
  bPrime = (2/len(x)) * sum
  return bPrime

###Gradiant Descent: Uses the Gradiant Descent method to determine the line of best fit.

learningRate=.000000001
a=0
aPast=0
b=0

#Use Gradiant Descent Formula to calculate final A and B after 10000 iterations.

for i in range(10000):
  aPast = a
  a += learningRate * aPrime(a, b, dataX, dataY)
  b += learningRate * bPrime(aPast, b, dataX, dataY)

#Display Values

loss = "%.2f" %loss(a, b, dataX, dataY)

print("")
print("")
print("Linear Regression Complete!")
print("")
print(f"The final loss is {loss}")
print("")

#Define xAxis and yAxis for new eqn.

xAxis = []
yAxis = []

z=0
for i in range(1000):
  xAxis.append(z)
  z+=1

for i in range(1000):
  yAxis.append(a*xAxis[i]+b) 

#Shorten a and b

m = "%.2f" % a
b = "%.2f" % b

#PLOT DATA

if b=="0.00":
  print(f"The equation is y={m}x")
elif b >= 0:
  print(f"The equation is y={m}x+{b}")
elif b < 0:
  print(f"The equation is y={m}x{b}")

print("")
print("")



plt.plot(dataX, dataY, 'ro', label="Given Points")
plt.axis([0, 20, 0, 100])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Linear Regression Chart - Small")
plt.grid(alpha=.4,linestyle='--')

if b=="0.00":
  plt.plot(xAxis, yAxis, label=f"y = {m}x")
elif b >= 0:
  plt.plot(xAxis, yAxis, label=f"y = {m}x+{b}")
elif b < 0:
  plt.plot(xAxis, yAxis, label=f"y = {m}x{b}")

plt.legend()
plt.show()

print("")
print("")
print("")

plt.plot(dataX, dataY, 'ro', label="Given Points")
plt.axis([0, 200, 0, 2000])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Linear Regression Chart - Fit")
plt.grid(alpha=.4,linestyle='--')

if b=="0.00":
  plt.plot(xAxis, yAxis, label=f"y = {m}x")
elif b >= 0:
  plt.plot(xAxis, yAxis, label=f"y = {m}x+{b}")
elif b < 0:
  plt.plot(xAxis, yAxis, label=f"y = {m}x{b}")
  
plt.legend()
plt.show()

print("")
print("")
print("")

plt.plot(dataX, dataY, 'ro', label="Given Points")
plt.axis([0, 200, 0, 10000])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("Linear Regression Chart - Mega")
plt.grid(alpha=.4,linestyle='--')

if b=="0.00":
  plt.plot(xAxis, yAxis, label=f"y = {m}x")
elif b >= 0:
  plt.plot(xAxis, yAxis, label=f"y = {m}x+{b}")
elif b < 0:
  plt.plot(xAxis, yAxis, label=f"y = {m}x{b}")
  
plt.legend()
plt.show()


```
