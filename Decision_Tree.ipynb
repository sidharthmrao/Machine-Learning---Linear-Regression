```
import numpy as np
import math

#FUNCTION: Get data from file into numpy matrix


def fileinstall():

  #Open File

  f = open("/content/drive/My Drive/AI Camp Data/Mushrooms (Decision Tree)/agaricus-lepiota.data", "r")

  #Append each line in file to 'given data'

  givenData = []                                                                                                                                                     
  for line in f:
    givenData.append(line.strip())

  #Append each character to list 'data'

  data=[]
  for index in range(len(givenData)):
    data.append(givenData[index].split(','))

  #Remove \n from end of list

  for index in range(len(data)):
    data[index][len(data[index])-1] = data[index][len(data[index])-1].rstrip('\n')

  data = np.asarray(data)
  
  return data

def datachange(data, bestval0):

  for instance in range(len(data[:,bestval0+1])):
    data[instance,bestval0+1] = '?'
  
  return data

#FUNCTION: Calculate how many of each item there are

def dataAmounts(data):
    dataAmounts = []
    index = 0
    y = []
    while(index<len(data[0,:])):
      b = {'#total' : 0 , '#types' : 0, '#vars' : ''}
      dataAmounts.append(b)
      z = 0
      while z<len(data[:,0])-1:
        dataAmounts[index]['#total'] +=1 
        if data[z,index] in dataAmounts[index]['#vars']:
          dataAmounts[index][data[z,index]] += 1
          z+=1
        else:
          dataAmounts[index]['#types'] += 1
          dataAmounts[index]['#vars'] += data[z,index]
          dataAmounts[index][data[z,index]] = 1
          z+=1
      index+=1
      y.append(list(dataAmounts[index-1].keys()))
    return dataAmounts, y

#Funtion: Calculates entropy for each set of data
def entropy(dataAmounts, keys):
  entropy = []
  individual = []
  index = 0
  while(index<len(data[0,:])):
    
    j = dataAmounts[index]['#total']
    individual = []
    typ = 0
    sumofp = 0
    while typ<(len(keys[index][3:])):
      
      z = dataAmounts[index][keys[index][typ+3]]/j
      l = math.log2(z)
      sumofp += z
      individual.append(z*l)
      typ+=1
    index+=1
    en = 0
    ran = 0
    while(ran<len(individual)):
      en+=(individual[ran])
      ran+=1
    entropy.append(en*-1)
  return entropy

def conditional(dataAmounts, keys):
  
  conditional = []
  conditioned = []

  index = 1
  while(index<len(data[1,:])):
    individual = []
    length = dataAmounts[index]['#types']
    types = dataAmounts[0]['#types']
    for var in range(len(keys[index][3:])):
      for t in range(types):
          individual.append([keys[index][var+3] , keys[0][t+3]])
    conditional.append(list(individual))
    index+=1
  
  for instance in range(len(conditional)):  #Instance = [[b = p], [b = e]] ...
    conditioned.append([])
    for eq in range(len(conditional[instance])):  #For i in range(len([b = p], [b = e])                        
      x = conditional[instance][eq][0]          
      y = conditional[instance][eq][1]
      conditioned[instance].append([x,y,0])
      for line in range(len(data[:,0])):
        if data[line, instance+1] == x :
          if data[line, 0] == y :
            conditioned[instance][eq][2] +=1

      if conditioned[instance][eq][2]>dataAmounts[instance+1][x]:
        conditioned[instance][eq][2] -=1
      conditioned[instance][eq].append(conditioned[instance][eq][2]/dataAmounts[instance+1][x])
      if conditioned[instance][eq][3]>0:
        conditioned[instance][eq].append(-1*(dataAmounts[instance+1][x]/dataAmounts[instance+1]['#total'])*conditioned[instance][eq][3]*math.log2(conditioned[instance][eq][3]))
      else:
        conditioned[instance][eq].append(0)

  return conditioned

def conditionalEntropy(conditioned):
  conditionalEntropies = []
  for index in range(len(conditioned)):
    conditionalEntropies.append(sum(conditioned[index][var][4] for var in range(len(conditioned[index]))))
  return conditionalEntropies

def bestsplit(entropy, conditionalEntropy):
  infogain = []
  for ent in range(len(conditionalEntropy)):
    infogain.append(entropy[0]-conditionalEntropy[ent])
  z = 0
  best = 0
  for i in range(len(infogain)):
    if infogain[i]>z :
      z = infogain[i]
      best = i
  return best, infogain[best]

def bestvar(conditioned, bestval):
  bestvar = 0
  z = 0
  for instance in range(len(conditioned[bestval])):
    if conditioned[bestval][instance][3] > z:
      z = conditioned[bestval][instance][3]
      bestvar = conditioned[bestval][instance][0]
      bestresult = conditioned[bestval][instance][1]
  return bestvar, bestresult, z

###################################################
data = fileinstall()

for i in range(5):
  dataAmounts0 = []
  keys0 = []
  entropy0 = []
  conditioned0 = []
  bestval0 = 0
  bestgain0 = 0
  bestvar0 = 0
  bestresut0 = 0
  conditionalProbability0 = 0
  best = ['',' 2nd',' 3rd',' 4th',' 5th']
  dataAmounts0, keys0 = dataAmounts(data)
  entropy0= entropy(dataAmounts0, keys0)
  conditioned0 = conditional(dataAmounts0, keys0)
  conditionalEntropy0 = conditionalEntropy(conditioned0)
  bestval0, bestgain0 = bestsplit(entropy0, conditionalEntropy0)
  bestvar0 , bestresult0, conditionalProbability0 = bestvar(conditioned0, bestval0)
  print(f"The{best[i]} split was on attribute {bestval0+1}, var {bestvar0} = {bestresult0}. The infogain of this attribute was {bestgain0}. The conditional probability of var {bestvar0} = {bestresult0} was {conditionalProbability0}.") 

  data = datachange(data, bestval0)
```
